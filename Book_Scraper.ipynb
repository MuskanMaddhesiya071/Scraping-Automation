```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from tqdm import tqdm
import numpy as np
```

### Book Details Scraping

```python
link='https://books.toscrape.com/catalogue/page-1.html'
```

```python
res=requests.get(link)
soup2=BeautifulSoup(res.text, 'html.parser')
```

```python
data=[]
for i in tqdm(range(1,51)):
    link='https://books.toscrape.com/catalogue/page-'+str(i)+'.html'
    res=requests.get(link)
    soup2=BeautifulSoup(res.text, 'html.parser')
    
    for sp in soup2.find_all('li', class_='col-xs-6 col-sm-4 col-md-3 col-lg-3'):
        book_link='https://books.toscrape.com/catalogue/'+sp.find_all('a')[-1].get('href')
        title=sp.find_all('a')[-1].get('title')
        image_link='https://books.toscrape.com/catalogue/'+sp.find('img').get('src')
        rating=sp.find('p').get('class')[-1]
        price=sp.find('p',class_='price_color').text[1:]
        instock=sp.find('p', class_='instock availability').text.strip()
    
        data.append([title, rating, price, instock,book_link,image_link ])
```

```python
pd.DataFrame(data, columns=['title', 'rating', 'price', 'instock', 'book_link', 'image_link'])
```
